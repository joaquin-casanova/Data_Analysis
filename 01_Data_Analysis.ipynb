# How to select specific rows and columns from a DataFrame?
![caption](imagen/03_subset_columns_rows1.svg)

specific N_PRO_VIV with P5G14 > 14

```python
specific_columns = data.loc[data["P5G14"] > 13, "N_PRO_VIV"]
```

```python
specific_columns.head()
```

In this case, a subset of both rows and columns is made one go and just selection brackets [] is not sufficient anymore. The loc/iloc 
operators ared required in front of the selection brackets [].  When using loc/iloc, the part before the comma is the  rows you want
and the part after the comma, is the columns you want to select

```python
data.iloc[9:25, 2:5], #rows and columns
```

We can assigned values to selected data

```python
data.iloc[0:5, 3] = "anonymous"
```

```python
data.head()
```

# REMEMBER
    * When selecting subsets of data, square brackets[] are used.
    * Inside these brackets, you can use a single column/row label, a list column/row labels, a slice of labels, a condition expresion or a colon.
    * Select specific rows and/or colums using loc when using the row and column name.
    * Select specific rows and/or columns using iloc when using the positions in the table.
    * You can assing new values to a selection based on loc/iloc
